```python
null_var = df.isnull().sum()/df.shape[0]*100
```

```python
# Here well we show the column name which have more than 18% missing data
drop_column = null_var[null_var >18].keys()
drop_column
```

```python
df2_drop_clm = df.drop(columns= drop_column)
```

```python
df2_drop_clm.shape
```

```python
sns.heatmap(df2_drop_clm.isnull())
```

```python
df3_drop_rows = df2_drop_clm.dropna()
```
